## 1. Finding the second largest element in an array

```python
def find_second_largest(l):  
    largest, second_largest =  l[0], l[0]
    for i in l:
        if i > largest:
            second_largest = largest
            largest = i
        elif i > second_largest and i != largest:
            second_largest = i

    return second_largest

print(find_second_largest([53,62,667,124,641,64]))
```

## 2. Union of 2 sorted arrays

```python
l1 = [1,1,2,3,4,5,6]
l2 = [2,3,4,4,5,6]
res = []
i, j = 0, 0
m, n = len(l1), len(l2)
while i+j < (m+n):
    if res and l1[i] == res[-1]:
        i += 1
    elif res and l2[j] == res[-1]:
        j += 1
    elif l1[i] < l2[j]:
        res.append(l1[i])
        i += 1
    elif l1[i] == l2[j]:
        res.append(l1[i])
        i, j = i+1, j+1
    else:
        res.append(l2[j])
        j += 1
print(res)
```

## 2. Intersection of 2 sorted arrays

```python
l1 = [1,1,2,3,4,5,6]
l2 = [2,3,4,4,5,6]
res = []
i, j = 0, 0
m, n = len(l1), len(l2)
while i+j < (m+n):
    if res and l1[i] == res[-1]:
        i += 1
    elif res and l2[j] == res[-1]:
        j += 1
    elif l1[i] == l2[j]:
        res.append(l1[i])
        i += 1
        j += 1
    elif l1[i] < l2[j]:
        i += 1
    else:
        j+=1
print(res)
```
