## 1. Finding the second largest element in an array

```python
def find_second_largest(l):  
    largest, second_largest =  l[0], l[0]
    for i in l:
        if i > largest:
            second_largest = largest
            largest = i
        elif i > second_largest and i != largest:
            second_largest = i

    return second_largest

print(find_second_largest([53,62,667,124,641,64]))
```
