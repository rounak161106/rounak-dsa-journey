```python
class Node:
    def __init__(self, data=None):
        self.prev = None
        self.data = data
        self.next = None

class doubly_linked_list:
    def __init__(self):
        self.head = None
        self.last = None

    def insert_end(self,v):
        new = Node(v)
        if self.head == None:
            self.head = new
            self.last = new
        else:
            self.last.next = new
            new.prev = self.last
            self.last = new

    def delete_end(self):
        if self.head == None:
            return
        
        if self.last == self.head:
            self.last = None
            self.head = None
            return
        
        self.last = self.last.prev
        self.last.next = None

    def __str__(self):
        l = []
        curr = self.head
        while curr !=None:
            l.append(curr.data)
            curr = curr.next
        return(str(l))
```

```python
dll = doubly_linked_list()
dll.insert_end(53)
dll.insert_end(63)
dll.insert_end(73)
print(dll)
dll.delete_end()
print(dll)
dll.delete_end()
print(dll)
```

```python
def DishPrepareOrder(order_list):
    count = {}
    for i in order_list:
        count[i] = count.get(i,0) + 1
        
    print(sorted(count, key=lambda x : (count[x], -x), reverse=True))

DishPrepareOrder([1009, 1001, 1005, 1004, 1008, 1002, 1003, 1006, 1007])
```

```python
def EvaluateExpression(exp):
    exp = exp.split(' ')
    stack = []
    for i in exp:
        if i.isdigit():
            stack.append(i)
        else:
            operator = i
            b = stack.pop()
            a = stack.pop()
            res = eval(f"{a}{operator}{b}")
            stack.append(res)
    print(float(stack[0]))

EvaluateExpression(input())
```

```python
class Node:
    def __init__(self, data = None):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def isempty(self):
        return self.head == None

    def append(self, data):
        if self.isempty():
            self.head = Node(data)
        else:
            temp = self.head
            while temp.next != None:
                temp = temp.next
            temp.next = Node(data)

    def delete(self, v):
        if self.isempty():
            return "List is empty."
        if self.head.data == v:
            self.head = self.head.next
            return
        temp = self.head
        temp1 = self.head
        while temp.next != None and temp.data != v:
            temp1 = temp
            temp = temp.next
        if temp.data == v:
            temp1.next = temp.next
            return 
        if temp.next == None:  
            return "Element doesn't exists"
    
    def reverse(self):
        prev = None
        curr = self.head
        while curr != None:
            next = curr.next
            curr.next = prev
            prev = curr
            curr = next

        self.head = prev

    def __str__(self):
        temp = self.head
        l = []
        while temp != None:
            l.append(temp.data)
            temp = temp.next
        return str(l)
```

```python
l = LinkedList()
l.append(10)
l.append(20)
l.append(30)
print(l)
l.reverse()
print(l)
```
