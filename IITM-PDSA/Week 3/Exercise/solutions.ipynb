```python
class Node:
    def __init__(self, data=None):
        self.prev = None
        self.data = data
        self.next = None

class doubly_linked_list:
    def __init__(self):
        self.head = None
        self.last = None

    def insert_end(self,v):
        new = Node(v)
        if self.head == None:
            self.head = new
            self.last = new
        else:
            self.last.next = new
            new.prev = self.last
            self.last = new

    def delete_end(self):
        if self.head == None:
            return
        
        if self.last == self.head:
            self.last = None
            self.head = None
            return
        
        self.last = self.last.prev
        self.last.next = None

    def __str__(self):
        l = []
        curr = self.head
        while curr !=None:
            l.append(curr.data)
            curr = curr.next
        return(str(l))
```

```python
dll = doubly_linked_list()
dll.insert_end(53)
dll.insert_end(63)
dll.insert_end(73)
print(dll)
dll.delete_end()
print(dll)
dll.delete_end()
print(dll)
```

```python
def DishPrepareOrder(order_list):
    count = {}
    for i in order_list:
        count[i] = count.get(i,0) + 1
        
    print(sorted(count, key=lambda x : (count[x], -x), reverse=True))

DishPrepareOrder([1009, 1001, 1005, 1004, 1008, 1002, 1003, 1006, 1007])
```
