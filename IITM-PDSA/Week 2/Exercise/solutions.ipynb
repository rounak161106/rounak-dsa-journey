```python
arr = [11,13,15,17]

n = len(arr)
low, high = 0, n - 1 
if n == 1:
    print(arr[0])
else:
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] > arr[mid+1]:
            print(arr[mid])
            break
        elif high-low == 1 and arr[high] > arr[low]:
            print(arr[high])
            break
        elif arr[mid] < arr[low]:
            high = mid-1
        elif arr[mid] > arr[low]:
            low = mid
```
