# Sorting Algorithms

## Selection Sort

In selection sort, we select the minimum element and swap it with the current value of i.
The complexity of this algorithm is O($n^2$) for every case best, worse and avg. because no matter what, whether the list is sorted or not, it scans entirely through the list.

```python
l = [13,53,12,53,1,56,64]
for i in range(len(l) - 1):
    mini = i 
    for j in range(i+1, len(l)):
        if l[j] < l[mini]:
            mini = j
    l[i], l[mini] = l[mini], l[i]

print(l)
```

## Insertion Sort
In insertion sort, we iterate over the list and for i th value, we will insert that to the correct position. We compare it with the previous elements to find its correct position. The complexity of this algorithm is O($n^2$) but in best case when the list is already sorted, it will take O(n) time.

## Merge Sort
complexity if O($n \log n$)

```python
def merge(l,r):
    m,n = len(l), len(r)
    c = []
    i,j,k = 0,0,0
    while k < m+n:
        if i == m:
            c.extend(r[j:])
            k = k + (n-j)
        elif j == n:
            c.extend(l[i:])
            k = k + (m-i)
        elif l[i] < r[j]:
            c.append(l[i])
            i,k = i+1,k+1
        else:
            c.append(r[j])
            j,k = j+1,k+1
    return c


def MergeSort(l):
    n = len(l)
    if n <= 1:
        return l
    left = MergeSort(l[:n//2])
    right = MergeSort(l[n//2:])
    b = merge(left,right)
    return b

print(MergeSort([12,5,8,3,15,7,2,10]))
```
