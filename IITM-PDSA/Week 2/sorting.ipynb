# Sorting Algorithms

## Selection Sort

In selection sort, we select the minimum element and swap it with the current value of i.
The complexity of this algorithm is O($n^2$) for every case best, worse and avg. because no matter what, whether the list is sorted or not, it scans entirely through the list.

```python
l = [13,53,12,53,1,56,64]
for i in range(len(l) - 1):
    mini = i 
    for j in range(i+1, len(l)):
        if l[j] < l[mini]:
            mini = j
    l[i], l[mini] = l[mini], l[i]

print(l)
```

## Insertion Sort
In insertion sort, we iterate over the list and for i th value, we will insert that to the correct position. We compare it with the previous elements to find its correct position. The complexity of this algorithm is O($n^2$)

```python
l = [13,53,12,53,1,56,64]
for i in range(1,len(l)):
    j = i
    while(j > 0 and l[j] < l[j-1]):
        l[j], l[j-1] = l[j-1],l[j]
        j -= 1

print(l)
```
