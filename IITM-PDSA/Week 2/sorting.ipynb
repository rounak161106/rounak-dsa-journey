# Sorting Algorithms

## Selection Sort

In selection sort, we select the minimum element and swap it with the current value of i.
The complexity of this algorithm is O($n^2$) for every case best, worse and avg. because no matter what, whether the list is sorted or not, it scans entirely through the list. It is not stable. It sorts the list in place.

```python
l = [13,53,12,53,1,56,64]
def selection_sort(l):
    for i in range(len(l) - 1):
        mini = i 
        for j in range(i+1, len(l)):
            if l[j] < l[mini]:
                mini = j
        l[i], l[mini] = l[mini], l[i]

    print(l)

selection_sort(l)
```

## Bubble Sort

```python
def bubble_sort(l):
    n = len(l)
    for i in range(n):
        swapped = False
        for j in range(n-i-1):
            if l[j] > l[j+1]:
                l[j+1], l[j] = l[j], l[j+1]
                swapped = True
        if not swapped:
            return l
    return l

print(bubble_sort([-32, -43, 80, 53, 26, 0, -632, 6, 63]))
```

## Insertion Sort
In insertion sort, we iterate over the list and for i th value, we will insert that to the correct position. We compare it with the previous elements to find its correct position. The complexity of this algorithm is O($n^2$) but in best case when the list is already sorted, it will take O(n) time. It is stable and it sorts the list in place.

```python
# l = [13,53,12,53,1,56,64]
# for i in range(1,len(l)):
#     j = i
#     while(j > 0 and l[j] < l[j-1]):
#         l[j], l[j-1] = l[j-1],l[j]
#         j -= 1

# print(l)

def Insert(l,v):
    if not l:
        return [v]
    if v > l[-1]:
        return l + [v]
    else:
        return Insert(l[:-1],v) + l[-1:]
    
def InsertionSort(l):
    if not l:
        return []
    l = Insert(InsertionSort(l[:-1]), l[-1])
    return l

print(InsertionSort([12,5,8,3,15,7,2,10]))
```

## Merge Sort
complexity is O($n \log n$). it is stable and it is not in place. 

```python
def merge(l,r):
    m,n = len(l), len(r)
    c = []
    i,j,k = 0,0,0
    while k < m+n:
        if i == m:
            c.extend(r[j:])
            k = k + (n-j)
        elif j == n:
            c.extend(l[i:])
            k = k + (m-i)
        elif l[i] < r[j]:
            c.append(l[i])
            i,k = i+1,k+1
        else:
            c.append(r[j])
            j,k = j+1,k+1
    return c


def MergeSort(l):
    n = len(l)
    if n <= 1:
        return l
    left = MergeSort(l[:n//2])
    right = MergeSort(l[n//2:])
    b = merge(left,right)
    return b

print(MergeSort([12,5,8,3,15,7,2,10]))
```

## Quick Sort
Its complexity is O($n \log n$) in average case and O($n^2$) in worst case. It is not stable and it sorts the list in place.
Here first we take first element as the pivot element. The strategy is simple, j is always incremented but i is incremented and swapped only if l[j] < pivot.
Average case complexity is O($n \log n$) because we are dividing the list into two halves and sorting them recursively. Worst case complexity is O($n^2$) because if the list is already sorted, then we will have to scan through the entire list for every element.

```python
def partition(l, lower, upper):
    pivot = l[lower]
    i = lower
    for j in range(lower+1, upper+1):
        if l[j] < pivot:
            i += 1
            l[i], l[j] = l[j], l[i]
    l[lower], l[i] = l[i], l[lower]
    return i 

def QuickSort(l, lower, upper):
    if lower < upper:
        pivot_pos = partition(l, lower, upper)
        QuickSort(l, lower, pivot_pos-1)
        QuickSort(l, pivot_pos+1, upper)
    return l

print(QuickSort([5, 6, 4, 3, 2, 1], 0, 5))
```
