# Week 2 : Searching

## Linear Search

```python
l = [35,264,62,623,74,23,74]
ele = 264
found = False
for i in l:
    if i == ele:
        print("Found at index", l.index(i))
        found = True
        break
if not found:
    print("Not found")
```

## Binary Search

```python
l = [45,64,74,78,79,89,90,91,92,93,94,95]
ele = 94
low, high = 0, len(l) - 1
found = False
while low <= high:
    mid = (low + high) // 2
    if l[mid] == ele:
        print("Found at index", mid)
        found = True
        break
    elif l[mid] < ele:
        low = mid + 1   
    else:
        high = mid - 1
if not found:
    print("Not found")
```

## Binary Search : using recursion

```python
def binarySearch(l, low, high, ele):
    if low > high:
        return "not found"
    
    mid = (low + high) // 2
    if (l[mid] == ele): 
        return mid

    elif(l[mid] > ele):
        return binarySearch(l, low, mid-1, ele)
    
    else:
        return binarySearch(l, mid+1, high, ele)

l = [45,64,74,78,79,89,90,91,92,93,94,95]
ele = 94
print(binarySearch(l, 0, len(l) - 1, ele))
```
