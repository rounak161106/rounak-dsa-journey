### A tree is a acyclic and minimally connected graph and has n-1 vertices. The Minimum Spanning Tree (MST) is defined only for connected, weighted, undirected graphs.

### 1. Prims Algorithm

```python
def prims_algo(ad_list):
    visited, edges = {}, []
    for i in ad_list:
        visited[i] = False
    
    visited[0] = True

    for i in range(len(ad_list) - 1):
        mini = float('inf')
        nextv, nexte = None, None
        for u in ad_list:
            for v, d in ad_list[u]:
                if visited[u] and not visited[v] and d < mini:  
                    mini, nextv, nexte = d, v, (u,v)
        if not nextv:

        visited[nextv] = True
        edges.append(nexte)
    
    return edges
```
