## Applications of BFS AND DFS

```python
class Queue:
    def __init__(self):
        self.queue = []

    def isempty(self):
        return self.queue == []

    def enqueue(self, v):
        self.queue.append(v)
    
    def dequeue(self):
        v = None
        if not self.isempty():
            v = self.queue[0]
            self.queue = self.queue[1:]
        return v

    def __str__(self):
        return str(self.queue)

def bfs_list(ad_list, v):
    queue = Queue()
    visited = {}
    for i in ad_list:
        visited[i] = False

    visited[v] = True
    queue.enqueue(v)

    while(not queue.isempty()):
        vertex = queue.dequeue()

        for i in ad_list[vertex]:
            if(not visited[i]):
                visited[i] = True
                queue.enqueue(i)

    return visited
```

### 1. Finding Connected Components

```python
def Components(ad_list):
    components = {}
    compid, seen = 0, 0
    for i in ad_list:
        components[i] = -1

    n = max(ad_list.keys())
    while seen <= n:
        start_vertex = min([i for i in components if components[i] == -1])
        visited = bfs_list(ad_list, start_vertex)
        for i in visited.keys():
            if visited[i]:
                seen += 1
                components[i] = compid
        compid += 1
    return components

print(Components({
    0: [1, 2],
    1: [0, 2],
    2: [0, 1, 3],
    3: [2],
    4: [5],
    5: [4],
    6: [],
    7: [8],
    8: [7, 9],
    9: [8]
}))
```

### Cyclic and Acyclic Graphs
1. Every non tree edge in an undirected graph contributes to a cycle.
2. There are back, forward and cross edges in a directed graph. The back edge always contribute to a cycle while the other 2 doesn't necessarily contribute to a cycle.
3. For forward edge u -> v, if the pre and post of v is contained in the pre and post of u, we say it is forward. and the opposite happens for backward edge. if the pre and post of u is contained in the pre and post of v, it back edge.
4. If the pre and post of u and v are disjoint, we say cross edge.

## Finding pre and post nos. using dfs

```python
visited, pre, post = {}, {}, {}
def init(ad_list):
    for i in ad_list:
        visited[i] = False
        pre[i], post[i] = -1, -1
    return

def dfs(ad_list, v, count):
    visited[v] = True
    pre[v] = count
    count += 1
    for i in ad_list[v]:
        if not visited[i]:
            count = dfs(ad_list, i, count)
    post[v] = count 
    count += 1
    return count
        
ad_list = {0: [1, 4],1: [0],4: [0,8,9],8: [4,9],9: [4,8]}
init(ad_list)
dfs(ad_list, 0, 0)
print(visited)
print(pre)
print(post)
```
