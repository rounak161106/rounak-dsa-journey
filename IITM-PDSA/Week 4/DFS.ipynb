```python
def neighbours(matrix, v):
    nbrs = []
    for col in range(len(matrix)):
        if matrix[v][col] == 1:
            nbrs.append(col)
    return nbrs
```

```python
ad_list = {}
edges = [(0, 1),(0, 2),(1, 2),(1, 3),(2, 4),(3, 5),(4, 5),(4, 6),(5, 7),(6, 7),(7, 8),(8, 9),(10, 11),(11, 12),(12, 13),(13, 14),(14, 15)] 
edges = edges + [(j,i) for (i,j) in edges]
matrix = [ [0 for i in range(16)] for j in range(16)]
for (i,j) in edges:
    matrix[i][j] = 1

for i in range(16):
    ad_list[i] = neighbours(matrix, i)

visited, parent = {}, {}
for i in ad_list:
    visited[i] = False
    parent[i] = -1

def dfs_list(Alist, v):
    visited[v] = True
    for i in Alist[v]:
        if(not visited[i]):
            parent[i] = v
            dfs_list(Alist, i)
    return

i = 1
dfs_list(ad_list, i)
print(i,"--->",visited)
print(parent)
```

## DFS implementation without recursion

```python
class Stack:
    def __init__(self):
        self.stack = []
    
    def isempty(self):
        return self.stack == []
    
    def push(self, v):
        self.stack.append(v)

    def pop(self):
        v = None
        if not self.isempty():
            v = self.stack.pop()
        return v
    
    def __str__(self):
        return str(self.stack)
```

```python
def dfs(ad_list, v):
    parent, visited = {}, {}
    for i in ad_list:
        visited[i] = False
        parent[i] = -1

    stack = Stack()
    stack.push(v)

    while(not stack.isempty()):
        v = stack.pop()
        if not visited[v]:
            visited[v] = True
            for i in ad_list[v]:
                if(not visited[i]):
                    parent[i] = v
                    stack.push(i)
    return visited, parent

ad_list = {
    0: [1, 2],
    1: [4, 5],
    2: [3],
    3: [6],
    4: [],
    5: [6],
    6: []
}

visited, parent = dfs(ad_list, 0)

print(visited)
print(parent)
```
