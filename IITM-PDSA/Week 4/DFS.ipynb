```python
def neighbours(matrix, v):
    nbrs = []
    for col in range(len(matrix)):
        if matrix[v][col] == 1:
            nbrs.append(col)
    return nbrs
```

```python
ad_list = {}
edges = [(0, 1),(0, 2),(1, 2),(1, 3),(2, 4),(3, 5),(4, 5),(4, 6),(5, 7),(6, 7),(7, 8),(8, 9),(10, 11),(11, 12),(12, 13),(13, 14),(14, 15)] 
edges = edges + [(j,i) for (i,j) in edges]
matrix = [ [0 for i in range(16)] for j in range(16)]
for (i,j) in edges:
    matrix[i][j] = 1

for i in range(16):
    ad_list[i] = neighbours(matrix, i)

visited, parent = {}, {}
for i in ad_list:
    visited[i] = False
    parent[i] = -1

def dfs_list(Alist, v):
    visited[v] = True
    for i in Alist[v]:
        if(not visited[i]):
            visited[i] = True
            parent[i] = v
            dfs_list(Alist, i)
    return

i = 1
dfs_list(ad_list, i)
print(i,"--->",visited)
print(parent)
```
