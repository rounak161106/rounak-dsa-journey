## BFS implementation using adjacency list and matrix

```python
class Queue:
    def __init__(self):
        self.queue = []

    def isempty(self):
        return self.queue == []

    def enqueue(self, v):
        self.queue.append(v)
    
    def dequeue(self):
        v = None
        if not self.isempty():
            v = self.queue[0]
            self.queue = self.queue[1:]
        return v

    def __str__(self):
        return str(self.queue)
```

```python
def bfs_list(ad_list, v):
    queue = Queue()
    visited = {}
    for i in ad_list:
        visited[i] = False

    visited[v] = True
    queue.enqueue(v)

    while(not queue.isempty()):
        vertex = queue.dequeue()

        for i in ad_list[vertex]:
            if(not visited[i]):
                visited[i] = True
                queue.enqueue(i)

    return visited
```

```python
def neighbours(matrix, v):
    nbrs = []
    for col in range(len(matrix)):
        if matrix[v][col] == 1:
            nbrs.append(col)
    return nbrs
```

```python
def bfs_matrix(matrix, v):
    queue = Queue()
    visited = {}
    for i in range(len(matrix)):
        visited[i] = False

    visited[v] = True
    queue.enqueue(v)

    while(not queue.isempty()):
        vertex = queue.dequeue()
        
        for i in neighbours(matrix, vertex):
            if(not visited[i]):
                visited[i] = True
                queue.enqueue(i)

    return visited
```

```python

ad_list = {}
edges = [(0, 1),(0, 2),(1, 2),(1, 3),(2, 4),(3, 5),(4, 5),(4, 6),(5, 7),(6, 7),(7, 8),(8, 9),(10, 11),(11, 12),(12, 13),(13, 14),(14, 15)] 
edges = edges + [(j,i) for (i,j) in edges]
matrix = [ [0 for i in range(16)] for j in range(16)]
for (i,j) in edges:
    matrix[i][j] = 1

for i in range(16):
    ad_list[i] = neighbours(matrix, i)
for i in ad_list.keys():    
    print(f"From vertex {i}, reachable vertices are => {bfs_list(ad_list, i)} - Adjacency list")
    print(f"From vertex {i}, reachable vertices are => {bfs_matrix(matrix, i)} - Adjacency matrix")
```
