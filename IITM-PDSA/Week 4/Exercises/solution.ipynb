```python
def findAllPaths(gList, source, destination):
    all_paths = []
    path = []

    def dfs(current):
        path.append(current)
        if current == destination:
            all_paths.append(path[:])
        else:
            for k in gList[current]:
                dfs(k)

        path.pop()
        return

    dfs(source)
    return all_paths

print(findAllPaths({1:[2,3], 2:[4], 3:[4], 4:[5,6], 5:[], 6:[]}, 1, 5))
```

```python
def subsets(nums):
    result = []
    path = []

    def dfs(start):
        result.append(path[:])           # every path IS a valid subset — save it immediately
        for i in range(start, len(nums)):
            path.append(nums[i])          # choose
            dfs(i + 1)                    # explore
            path.pop()                    # un-choose

    dfs(0)
    return result
subsets([1,2])
```
